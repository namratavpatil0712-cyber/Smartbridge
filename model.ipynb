```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
url = "https://raw.githubusercontent.com/namratavpatil0712-cyber/Smartbridge/main/hypertension.csv"
df = pd.read_csv(url)
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()

plt.figure(figsize=(5,5))
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Gender Pie Chart")
plt.ylabel('')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Stages', data=df)
plt.title("Hypertension Stage Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
corr = df[['Systolic', 'Diastolic']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation between Systolic and Diastolic")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Severity', hue='TakeMedication', data=df)
plt.title("Medication vs Severity")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Age', hue='Stages', data=df)
plt.title("Age vs Hypertension Stage")
plt.show()
```

```python
sns.pairplot(df[['Systolic', 'Diastolic', 'Stages']], hue='Stages')
plt.show()
```
